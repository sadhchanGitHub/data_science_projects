```python
"""
check the file after split combine load_combine_save_categories_for_TrainValTest_dask.py

"""
import numpy as np

# Paths to combined datasets
train_data_path = "../data/preprocessed_output_data/train.npy"
train_masks_path = "../data/preprocessed_output_data/train_masks.npy"
val_data_path = "../data/preprocessed_output_data/val.npy"
val_masks_path = "../data/preprocessed_output_data/val_masks.npy"
test_data_path = "../data/preprocessed_output_data/test.npy"
test_masks_path = "../data/preprocessed_output_data/test_masks.npy"

# Load data
x_train = np.load(train_data_path, mmap_mode='r')
masks_train = np.load(train_masks_path, mmap_mode='r')
x_val = np.load(val_data_path, mmap_mode='r')
masks_val = np.load(val_masks_path, mmap_mode='r')
x_test = np.load(test_data_path, mmap_mode='r')
masks_test = np.load(test_masks_path, mmap_mode='r')

# Check shapes
assert x_train.shape[0] == masks_train.shape[0], f"Train mismatch: {x_train.shape[0]} images, {masks_train.shape[0]} masks"
assert x_val.shape[0] == masks_val.shape[0], f"Validation mismatch: {x_val.shape[0]} images, {masks_val.shape[0]} masks"
assert x_test.shape[0] == masks_test.shape[0], f"Test mismatch: {x_test.shape[0]} images, {masks_test.shape[0]} masks"

print("All checks passed! Shapes match:")
print(f"Train: {x_train.shape[0]} images, {masks_train.shape[0]} masks")
print(f"Validation: {x_val.shape[0]} images, {masks_val.shape[0]} masks")
print(f"Test: {x_test.shape[0]} images, {masks_test.shape[0]} masks")
```
