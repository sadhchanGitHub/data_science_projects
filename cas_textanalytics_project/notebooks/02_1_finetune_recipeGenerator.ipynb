```python

# Log GPU usage
def log_gpu_usage():
    with open("gpu_usage_log.txt", "a") as f:
        while True:
            gpu_status = subprocess.check_output("nvidia-smi", shell=True).decode()
            f.write(gpu_status)
            f.write("\n" + "-"*80 + "\n")
            time.sleep(300)  # Log every 5 minutes
```

```python
#Setup Logging Callback for Training
#Add the custom callback for logging training progress:
                       
class CustomLoggingCallback(TrainerCallback):
    def on_log(self, args, state, control, logs=None, **kwargs):
        if logs:
            with open("training_log.txt", "a") as f:
                f.write(f"Step {state.global_step}: {logs}\n")
```

```python
from evaluate import load

# Load BLEU metric
bleu_metric = load("bleu")

def compute_metrics(eval_pred):
    predictions, labels = eval_pred

    # Decode predictions and labels
    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    # Prepare references as a list of lists
    decoded_labels = [[label] for label in decoded_labels]

    # Compute BLEU score
    result = bleu_metric.compute(predictions=decoded_preds, references=decoded_labels)
    return {"bleu": result["score"]}
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel, Trainer, TrainingArguments

# Load tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")

# Add padding token to tokenizer
tokenizer.pad_token = tokenizer.eos_token

# Tokenize the dataset
def tokenize_function(examples):
    combined_texts = [
        ingredient + " " + direction
        for ingredient, direction in zip(examples["ingredients"], examples["directions"])
    ]
    return tokenizer(combined_texts, truncation=True, padding=True)

tokenized_dataset = hf_dataset.map(tokenize_function, batched=True)
```

```python
def write_generated_recipes_to_csv(output_file, recipes, append=False):
    """
    Save a list of recipes to a CSV file.
    
    Parameters:
        output_file (str): Path to the output CSV file.
        recipes (list): List of recipe dictionaries.
        append (bool): Whether to append to the file if it exists. Defaults to False.
    """
    # Ensure the directory exists
    os.makedirs(os.path.dirname(output_file), exist_ok=True)
    
    try:
        # Convert recipes to a DataFrame
        df = pd.DataFrame(recipes)
        
        # Save to CSV, either appending or overwriting
        if append and os.path.exists(output_file):
            df.to_csv(output_file, mode='a', header=False, index=False)
        else:
            df.to_csv(output_file, index=False)
        
        print(f"Recipes saved to {output_file}")
    except Exception as e:
        print(f"Error saving recipes to {output_file}: {e}")
```

```python
import itertools
import pandas as pd
import os

def post_process_recipe(recipe, input_ingredients):
    """
    Cleans and formats the generated recipe for better readability.
    - Deduplicates sentences.
    - Ensures proper structure for ingredients and instructions.
    """
    import re

    # Split recipe into parts
    parts = recipe.split("]")

    # Extract instructions if they exist
    if len(parts) > 1:
        instructions = parts[1]
    else:
        instructions = recipe

    # Clean up instructions: remove duplicates and fix formatting
    steps = re.split(r'(?<=[.!?])\s+', instructions)  # Split by sentences
    seen = set()
    cleaned_steps = [step for step in steps if step not in seen and not seen.add(step)]

    # Combine cleaned steps
    final_instructions = "\n".join(cleaned_steps)

    # Format output
    return f"Ingredients: {input_ingredients}\n\nInstructions:\n{final_instructions}"

```

```python
import re

def clean_duplicates(recipe):
    # Split on sentence boundaries using regex
    sentences = re.split(r'(?<=[.!?])\s+', recipe)
    seen = set()
    cleaned = [sentence for sentence in sentences if not (sentence in seen or seen.add(sentence))]
    return " ".join(cleaned)
```

```python
import re

def clean_full_recipe(recipe):
    """
    Cleans the recipe by removing duplicate sentences and ensuring consistent formatting.
    Handles edge cases where the recipe format may be inconsistent.
    """
    # Split recipe into parts based on "]" if it exists (ingredients and instructions)
    parts = recipe.split("]")
    
    if len(parts) > 1:
        # Assume everything after "]" contains instructions
        instructions = clean_duplicates("]".join(parts[1:]))
        return f"{parts[0]}] {instructions}"
    else:
        # If no "]" exists, apply sentence-level deduplication to the whole recipe
        return clean_duplicates(recipe)

```

```python
def generate_recipes_batch(dataset, generator, output_file, max_length=200, save_interval=10):
    """
    Generate recipes in batch for all parameter combinations.
    Periodically save intermediate results to avoid data loss.
    """
    recipes = []
    saved_count = 0  # Keep track of saved recipes

    for idx, row in dataset.iterrows():
        try:
            # Extract parameters and input ingredients
            input_ingredients = row["input_ingredients"]
            top_k = row["top_k"]
            top_p = row["top_p"]
            temperature = row["temperature"]

            # Generate a recipe with the current parameters
            recipe = generator(
                input_ingredients,
                max_length=max_length,
                num_return_sequences=1,
                do_sample=True,
                top_k=top_k,
                top_p=top_p,
                temperature=temperature,
            )[0]["generated_text"]

            # Post-process the generated recipe
            cleaned_recipe = post_process_recipe(recipe, input_ingredients)
            print(cleaned_recipe)

            # Append the recipe and parameters to the results
            recipes.append({
                "input_ingredients": input_ingredients,
                "top_k": top_k,
                "top_p": top_p,
                "temperature": temperature,
                "recipe": cleaned_recipe
            })

            # Save intermediate results every `save_interval` iterations
            if (idx + 1) % save_interval == 0:
                write_generated_recipes_to_csv(output_file, recipes[saved_count:], append=True)
                saved_count = len(recipes)  # Update saved count

        except Exception as e:
            print(f"Error generating recipe for {row}: {e}")
            recipes.append({
                "input_ingredients": input_ingredients,
                "top_k": top_k,
                "top_p": top_p,
                "temperature": temperature,
                "recipe": "Error generating recipe"
            })

    # Final save to ensure all recipes are written to the file
    write_generated_recipes_to_csv(output_file, recipes[saved_count:], append=True)

    return recipes
```
